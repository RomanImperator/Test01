```python
# @title 8. AVVIO WebApp Streamlit + tunnel Ngrok / Cloudflare
# @markdown ⚠️ ATTENZIONE!⚠️
# @markdown
# @markdown **Per eseguire i file corretti**, verificare e/o cambiare il valore della varialbile PROJECT_SRC, presente nel sorgente della cella, con il path
# @markdown
# @markdown "/content/drive/MyDrive/Near_Miss_Scuola/src"
# @markdown
# @markdown o comunque dove sono contenuti i file sorgenti da runnare. La cartella "src_collaudo" **non è condivisa** e all'interno sono contenuti i file in fase di collaudo
# @markdown
# @markdown ℹ️ Avvia la WebApp e apre un tunnel pubblico usando:
# @markdown - **Ngrok** se `TUNNEL_PROVIDER = "ngrok"` (richiede token),
# @markdown - **Cloudflare Quick Tunnel** se `TUNNEL_PROVIDER = "cloudflare"` (senza account).
# @markdown
# @markdown Il provider si imposta in `config_v06_03.py`.

import os, sys, time, socket, subprocess, re, urllib.request, urllib.error, random, importlib

# --------------------------------------------------------------------
# 1. Percorsi base del progetto
# --------------------------------------------------------------------
PROJECT_SRC = "/content/drive/MyDrive/Near_Miss_Scuola/src"
APP_PATH    = f"{PROJECT_SRC}/Near_Miss-SL_v06_03.py"
RUN_LOG     = "/content/streamlit_run.log"

if PROJECT_SRC not in sys.path:
    sys.path.insert(0, PROJECT_SRC)
os.environ["PYTHONPATH"] = PROJECT_SRC

# --------------------------------------------------------------------
# 2. Carico la config “pulita”
# --------------------------------------------------------------------
if "config_v06_03" in sys.modules:
    del sys.modules["config_v06_03"]

import config_v06_03 as cfg
cfg = importlib.reload(cfg)

from config_v06_03 import TUNNEL_PROVIDER, STREAMLIT_PORT
PORT     = STREAMLIT_PORT
PROVIDER = (TUNNEL_PROVIDER or "ngrok").strip().lower()
NGROK_TOKEN = os.getenv("NGROK_TOKEN", "").strip()

print("🚀 Avvio Streamlit su porta:", PORT)
print("➡️ TUNNEL_PROVIDER:", PROVIDER)

def wait_for_port(host: str, port: int, timeout: int = 90) -> bool:
    deadline = time.time() + timeout
    while time.time() < deadline:
        try:
            with socket.create_connection((host, port), timeout=1):
                return True
        except OSError:
            time.sleep(1)
    return False

# --------------------------------------------------------------------
# 3. Chiudo eventuali processi residui
# --------------------------------------------------------------------
subprocess.run(["pkill", "-f", "streamlit"],   check=False)
subprocess.run(["pkill", "-f", "ngrok"],       check=False)
subprocess.run(["pkill", "-f", "cloudflared"], check=False)

# --------------------------------------------------------------------
# 4. Avvio Streamlit in background
# --------------------------------------------------------------------
out_file = open(RUN_LOG, "w")
st_env = os.environ.copy()
st_env["PYTHONPATH"] = PROJECT_SRC

subprocess.Popen(
    ["streamlit", "run", APP_PATH, "--server.port", str(PORT), "--server.address", "0.0.0.0", "--server.headless", "true"],
    stdout=out_file, stderr=subprocess.STDOUT, text=True, env=st_env
)

print(f"⏳ Attendo che Streamlit risponda su 127.0.0.1:{PORT}…")
if not wait_for_port("127.0.0.1", PORT):
    print("❌ Streamlit non è partito nei tempi previsti. Ultimi log:")
    try: print(open(RUN_LOG).read()[-500:])
    except: pass
    sys.exit(1)
print("✅ Streamlit è in ascolto.")

# --------------------------------------------------------------------
# 5. Funzioni helper per i tunnel
# --------------------------------------------------------------------
def open_ngrok(port: int) -> str:
    print("⬇️ Preparo Ngrok…")
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "pyngrok"], check=True)
    from pyngrok import ngrok
    if not NGROK_TOKEN or NGROK_TOKEN == "MANCANTE":
        raise EnvironmentError("❌ NGROK_TOKEN mancante nel file .env")
    try:
        ngrok.set_auth_token(NGROK_TOKEN)
        return ngrok.connect(port).public_url
    except Exception as e:
        raise RuntimeError(f"Fallimento tunnel Ngrok: {e}")

def open_cloudflare(port: int) -> str:
    print("⬇️ Scarico/Avvio Cloudflare Quick Tunnel…")
    if not os.path.exists("cloudflared"):
        os.system("wget -q https://github.com/cloudflare/cloudflared/releases/latest/download/cloudflared-linux-amd64 -O cloudflared")
        os.system("chmod +x cloudflared")
    logf = "cloudflared.log"
    try:
        if os.path.exists(logf): os.remove(logf)
    except: pass

    cmd = (
        f"./cloudflared tunnel "
        f"--no-autoupdate "
        f"--protocol http2 "
        f"--edge-ip-version 4 "
        f"--url http://localhost:{port} "
        f"> {logf} 2>&1 &"
    )
    os.system(cmd)

    print("⏳ Attendo l'URL pubblico (max 40s)…")
    url = None
    for i in range(80):
        time.sleep(0.5)
        try:
            txt = open(logf, "r").read()
            m = re.findall(r"https://[-0-9a-z]+\.trycloudflare\.com", txt)
            if m:
                url = m[-1]
                break
        except: pass

    if not url:
        print("❌ Cloudflare non ha generato l'URL nei tempi previsti.")
        try: print("--- Ultimi Log Cloudflare ---\n", open(logf).read()[-600:])
        except: pass
    return url

# --------------------------------------------------------------------
# 6. Apertura tunnel
# --------------------------------------------------------------------
public_url = None
try:
    if PROVIDER == "cloudflare":
        public_url = open_cloudflare(PORT)
        if not public_url:
            print("⚠️ Cloudflare fallito, passo al fallback Ngrok…")
            public_url = open_ngrok(PORT)
    else:
        public_url = open_ngrok(PORT)

    time.sleep(0.5)

    if public_url:
        print("\n✅ Tunnel attivo!")
        print("🔗 Apri questo URL pubblico:", public_url)
    else:
        print("❌ Impossibile attivare un tunnel valido.")
except Exception as e:
    print(f"\n❌ Errore durante l'apertura del tunnel: {e}")
```

```python
# @title 10. Verifica automatica del sistema (Test)
# @markdown Questa cella esegue una serie di controlli automatici
# @markdown per verificare che i componenti principali della WebApp
# @markdown funzionino correttamente:
# @markdown - Riconoscimento del tipo di file CSV (predefinito o generico)
# @markdown - Collegamento al Google Sheet
# @markdown - Caricamento dati con piano B automatico
# @markdown - Lettura di file CSV con diversi separatori (; , Tab)
# @markdown
# @markdown ⚠️ ATTENZIONE!⚠️
# @markdown
# @markdown **Per eseguire i file corretti**, verificare e/o cambiare il valore della varialbile PROJECT_SRC, presente nel sorgente della cella, con il path
# @markdown
# @markdown "/content/drive/MyDrive/Near_Miss_Scuola/src"
# @markdown
# @markdown o comunque dove sono contenuti i file sorgenti da runnare. La cartella "src_collaudo" **non è condivisa** e all'interno sono contenuti i file in fase di collaudo
# @markdown

import os, sys, csv, io
from unittest.mock import patch, MagicMock
import pandas as pd

# Aggiungo la cartella sorgenti al percorso di ricerca
PROJECT_SRC = "/content/drive/MyDrive/Near_Miss_Scuola/src/v06_03"
if PROJECT_SRC not in sys.path:
    sys.path.insert(0, PROJECT_SRC)

# Contatori risultati
n_ok = 0
n_ko = 0

def test_ok(nome):
    global n_ok
    n_ok += 1
    print(f"  ✅ {nome}")

def test_ko(nome, errore):
    global n_ko
    n_ko += 1
    print(f"  ❌ {nome}: {errore}")

# Mock di Streamlit (non disponibile come widget in questa cella)
mock_st = MagicMock()
mock_st.session_state = {}

print("="*60)
print("CONTROLLI AUTOMATICI — Near Miss Scuola v06_03.5")
print("="*60)

# --- BLOCCO 1: Importazione moduli ---
print("\n📦 Importazione moduli...")
try:
    with patch.dict('sys.modules', {'streamlit': mock_st}):
        import importlib
        if 'Utils_NearMiss_v06_03' in sys.modules:
            del sys.modules['Utils_NearMiss_v06_03']
        if 'config_v06_03' in sys.modules:
            del sys.modules['config_v06_03']
    import config_v06_03 as cfg
    test_ok("config_v06_03 importato")
except Exception as e:
    test_ko("Importazione config_v06_03", e)

try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        import Utils_NearMiss_v06_03 as utils
    test_ok("Utils_NearMiss_v06_03 importato")
except Exception as e:
    test_ko("Importazione Utils_NearMiss_v06_03", e)

# --- BLOCCO 2: Riconoscimento tipo CSV ---
print("\n📄 Controllo riconoscimento tipo di file CSV...")
try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        # CSV predefinito (Near Miss)
        cols_predef = list(cfg.PREDEF_X_CHOICES.values())
        df_predef = pd.DataFrame(columns=cols_predef)
        assert utils.is_csv_predef(df_predef), "Doveva riconoscere il tracciato Near Miss"
        test_ok("CSV predefinito riconosciuto correttamente")
except Exception as e:
    test_ko("Riconoscimento CSV predefinito", e)

try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        # CSV generico (non Near Miss)
        df_gen = pd.DataFrame(columns=["Nome", "Cognome", "Voto"])
        assert not utils.is_csv_predef(df_gen), "Non doveva riconoscerlo come predefinito"
        test_ok("CSV generico riconosciuto correttamente")
except Exception as e:
    test_ko("Riconoscimento CSV generico", e)

# --- BLOCCO 3: Collegamento Google Sheet ---
print("\n🔌 Controllo collegamento Google Sheet...")
try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        # Simulo un server che risponde OK
        with patch('urllib.request.urlopen') as mock_url:
            mock_resp = MagicMock()
            mock_resp.status = 200
            mock_resp.__enter__ = MagicMock(return_value=mock_resp)
            mock_resp.__exit__ = MagicMock(return_value=False)
            mock_url.return_value = mock_resp
            result = utils.google_sheet_available(timeout_sec=1)
            assert result == True, "Doveva rispondere True"
        test_ok("Google Sheet raggiungibile (simulato)")
except Exception as e:
    test_ko("Verifica Google Sheet OK", e)

try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        # Simulo un errore di rete
        with patch('urllib.request.urlopen', side_effect=Exception('Errore rete')):
            result = utils.google_sheet_available(timeout_sec=1)
            assert result == False, "Doveva rispondere False"
        test_ok("Google Sheet non raggiungibile (gestito)")
except Exception as e:
    test_ko("Verifica Google Sheet errore", e)

# --- BLOCCO 4: Caricamento dati con piano B ---
print("\n📥 Controllo caricamento dati e piano B...")
try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        with patch('Utils_NearMiss_v06_03.carica_da_google_sheet', return_value=None):
            local_df = pd.DataFrame({"Test": ["Valore"]})
            with patch('pandas.read_csv', return_value=local_df):
                df, fonte = utils.carica_dati("google")
                assert fonte == "CSV locale/predefinito", f"Valore atteso errato: {fonte}"
        test_ok("Piano B: quando Google Sheet non funziona, usa il file locale")
except Exception as e:
    test_ko("Caricamento con piano B", e)

try:
    with patch('Utils_NearMiss_v06_03.st', mock_st):
        local_df = pd.DataFrame({"Col": ["Val"]})
        with patch('pandas.read_csv', return_value=local_df):
            df, fonte = utils.carica_dati("csv")
            assert fonte == "CSV locale/predefinito"
        test_ok("Caricamento file locale diretto")
except Exception as e:
    test_ko("Caricamento CSV locale", e)

# --- BLOCCO 5: Lettura CSV con separatori diversi ---
print("\n📊 Controllo lettura CSV con diversi separatori...")

# Test punto e virgola (tipico italiano)
try:
    testo_pv = "Nome;Cognome;Eta\nMario;Rossi;30\nLuigi;Bianchi;25"
    df_pv = pd.read_csv(io.StringIO(testo_pv), sep=";", engine="python")
    assert len(df_pv.columns) == 3, f"Attese 3 colonne, trovate {len(df_pv.columns)}"
    assert len(df_pv) == 2, f"Attese 2 righe, trovate {len(df_pv)}"
    test_ok("Separatore punto e virgola (;)")
except Exception as e:
    test_ko("CSV con punto e virgola", e)

# Test virgola (standard internazionale)
try:
    testo_v = "Name,Surname,Age\nJohn,Doe,28\nJane,Smith,32"
    df_v = pd.read_csv(io.StringIO(testo_v), sep=",", engine="python")
    assert len(df_v.columns) == 3
    test_ok("Separatore virgola (,)")
except Exception as e:
    test_ko("CSV con virgola", e)

# Test tabulazione
try:
    testo_t = "Campo1\tCampo2\tCampo3\nA\tB\tC"
    df_t = pd.read_csv(io.StringIO(testo_t), sep="\t", engine="python")
    assert len(df_t.columns) == 3
    test_ok("Separatore tabulazione (Tab)")
except Exception as e:
    test_ko("CSV con tabulazione", e)

# Test rilevamento automatico separatore
try:
    testo_auto = "Nome;Cognome;Eta\nMario;Rossi;30"
    dialetto = csv.Sniffer().sniff(testo_auto[:8192], delimiters=";,\t|")
    assert dialetto.delimiter == ";", f"Rilevato {dialetto.delimiter} invece di ;"
    test_ok("Rilevamento automatico del separatore")
except Exception as e:
    test_ko("Rilevamento automatico", e)

# Test: separatore sbagliato produce una sola colonna
try:
    testo_err = "Nome;Cognome;Eta\nMario;Rossi;30"
    df_err = pd.read_csv(io.StringIO(testo_err), sep=",", engine="python")
    assert len(df_err.columns) == 1, "Con separatore sbagliato doveva avere 1 colonna"
    test_ok("Avviso separatore sbagliato (1 sola colonna)")
except Exception as e:
    test_ko("Verifica separatore sbagliato", e)

# --- RIEPILOGO FINALE ---
print("\n" + "="*60)
totale = n_ok + n_ko
if n_ko == 0:
    print(f"🎉 Tutti i {totale} controlli superati!")
else:
    print(f"⚠️ {n_ok}/{totale} controlli superati, {n_ko} con problemi.")
print("="*60)
```
